```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
league_winners = pd.read_csv('data/league_winners.csv')
league_winners
```

```python
leagues = list(league_winners.columns)
for league in leagues:
    print("{} winners in last decade:\n{}\n".format(league, league_winners[league].value_counts()))
```

League 1 has most number of different winners, 6. But after new owners, Paris Saint-Germain have dominated the league by huge margin. Premier League has 4 different winners but most evenly distributed. Premier League has most contenders for League title at the moment. But Ligue 1 has seen most different winners in last decade. Now lets check how tight is the race for title by points.

```python
league_dfs = {}
for league in leagues:
    league_dfs[league] = pd.read_csv('data/'+league+'.csv')
```

```python
plt.figure(figsize=(12,6))
for league in leagues:
    plt.plot(league_dfs[league].apply(lambda x: x[0]-x[1]), label=league)
plt.legend()
plt.xlabel('Year')
plt.ylabel('Point Difference')
plt.title('Point Difference between Winner and Runner-Up')
```

```python
plt.figure(figsize=(12,6))
for league in leagues:
    plt.plot(league_dfs[league].apply(lambda x: x[0]-x[3]), label=league)
plt.legend()
plt.xlabel('Year')
plt.ylabel('Point Difference')
plt.title('Point Difference between Winner and 4th Team')
```

```python
plt.figure(figsize=(12,6))
for league in leagues:
    plt.plot(league_dfs[league].apply(lambda x: x[0]-x[9]), label=league)
plt.legend()
plt.xlabel('Year')
plt.ylabel('Point Difference')
plt.title('Point Difference between Winner and 10th Team')
```

```python
plt.figure(figsize=(12,6))
for league in leagues:
    plt.plot(league_dfs[league].apply(lambda x: x[0]-x.mean()), label=league)
plt.legend()
plt.xlabel('Year')
plt.ylabel('Point Difference')
plt.title('Point Difference between Winner and Mean of Points')
```

```python
league_dfs['ligue1']
```

```python
print("Average difference of points between winner and median (10th) team for different leagues:\n\n")
for league in leagues:
    print("{}: {}".format(league, league_dfs[league].apply(lambda x: x[0]-x[9]).mean()))
```

```python
print("Average difference of points between winner and runner-up for different leagues:\n\n")
for league in leagues:
    print("{}: {}".format(league, league_dfs[league].apply(lambda x: x[0]-x[1]).mean()))
```

```python
print("Average difference of points between winner and 4th team for different leagues:\n\n")
for league in leagues:
    print("{}: {}".format(league, league_dfs[league].apply(lambda x: x[0]-x[3]).mean()))
```

It is a tight race between winner and runner up in La Liga, mainly because of Real Madrid and Barcelona. League 1 has lower average point difference between winner and median (10th) placed team. But if we see the difference between winner and 4th team (Champions League spots), Premier League registers minimum difference of points on average. Usually top 4 or 5 teams challenge for League Title. Manchester United, Manchester City, Chelsea and occasionaly Liverpool and Tottenham are giving strong competition for title in EPL. Thus, it can be concluded that Premier League is the most competitive league. Had it not been for new owners of PSG in France, Ligue 1 would have been the most competitive league by far.

La Liga, Serie A and Ligue 1 are witnessing increasing gap between top and average teams.

## Visualization

```python
import plotly
import plotly.graph_objs as go
from plotly import tools
```

```python
plotly.offline.init_notebook_mode(connected=True)
```

```python
# Get Formatted Data

first_tenth = [] #list of average point difference between winner (first) and tenth placed team
first_second = []
first_fourth = []

for league in leagues:
    first_tenth.append(league_dfs[league].apply(lambda x: x[0]-x[9]).mean())
    first_second.append(league_dfs[league].apply(lambda x: x[0]-x[1]).mean())
    first_fourth.append(league_dfs[league].apply(lambda x: x[0]-x[3]).mean())
```

```python
viz_df = pd.DataFrame({'league':leagues, 'first_tenth':first_tenth, 'first_second':first_second, 'first_fourth':first_fourth})
viz_df
```

```python
viz_df['league'] = viz_df['league'].apply(lambda x: ' '.join(x.split('_')).title())
viz_df
```

```python
viz_df.sort_values(by='first_tenth')
```

```python
viz_df = viz_df.sort_values(by='first_tenth')
data = [go.Bar(x=viz_df['league'], y=viz_df['first_tenth'], width=0.5, marker=dict(color='#94efcf',line=dict(color='#158765', width=1.5)), text=viz_df['first_tenth'], textposition='auto')]
layout = {
    'title' : 'Average Difference of Points Between Winner and Median (10th) Team',
    'xaxis' : {'title':'League', 'showgrid':False},
    'yaxis' : {'title':'Point Difference Between Winner and Tenth Team', 'showgrid':False}
}
plotly.offline.iplot({'data':data, 'layout':layout})
```

```python
# to save svg image of above graph
#plotly.offline.plot({'data' : data, 'layout' : layout}, show_link=False, image='svg')
```

```python
plotly.offline.plot({'data' : data, 'layout' : layout}, validate=False, show_link=False, include_plotlyjs=False, output_type='div')
```

```python
viz_df = viz_df.sort_values(by='first_second')
data = [go.Bar(x=viz_df['league'], y=viz_df['first_second'], width=0.5, marker=dict(color='#8cd7ea',line=dict(color='#1489a8', width=1.5)), text=viz_df['first_second'], textposition='auto')]
layout = {
    'title' : 'Average Difference of Points Between Winner and Runner-up',
    'xaxis' : {'title':'League', 'showgrid':False},
    'yaxis' : {'title':'Point Difference Between Winner and Runner-up', 'showgrid':False}
}
plotly.offline.iplot({'data':data, 'layout':layout})
```
